# Análise de Vendas em Python

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
dados = {
    'data': ['2024-01-05','2024-01-10','2024-02-03','2024-02-15','2024-03-01'],
    'produto': ['Notebook','Mouse','Notebook','Teclado','Mouse'],
    'quantidade': [2,5,1,3,4],
    'preco': [3500,80,3500,150,80]
}

df = pd.DataFrame(dados)
df
```

```
df.info()
```

```
df['faturamento'] = df['quantidade'] * df['preco']
df
```

```
df['faturamento'].sum()
```

```
df.groupby('produto')['quantidade'].sum().sort_values(ascending=False)
```

```
df['data'] = pd.to_datetime(df['data'])

vendas_mes = df.groupby(df['data'].dt.month)['faturamento'].sum()
vendas_mes
```
